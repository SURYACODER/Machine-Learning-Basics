```python
import numpy as np
import matplotlib.pyplot as plt
```

**Dot product** 

```python
a = np.array([2,3])
b = np.array([4,4])
a_dot_b = np.dot(a,b)
a_dot_b
```

```python
c = np.array([11,23,45])
d = np.array([23,34,56])
e = np.dot(c,d)
e
```

**Cross Product**

```python
axb = np.cross(a,b)
axb
```

```python
cxd = np.cross(c,d)
cxd
```

**Projection of vectors**




```python
a = np.array([2,5])
v = np.array([8,-6])
# magnitude of v
mag_v = np.sqrt(sum(v**2))
print(mag_v)
```

```python
proj_a_on_v = (np.dot(a,v)/mag_v**2)*v
proj_a_on_v
```

```python
plt.quiver(0,0,2,5,scale_units='xy',angles='xy',scale=1,color='b' )
plt.quiver(0,0,8,-6,scale_units='xy',angles='xy',scale=1,color='r' )
plt.quiver(0,0,-1.12,0.84,scale_units='xy',angles='xy',scale=1,color='y' )
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```
