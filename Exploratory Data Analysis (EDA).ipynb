Haberman Dataset


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/content/haberman.csv')
data.head()
```

```python
df =pd.DataFrame()
df['age'] = data['30']
df['year'] = data['64']
df['nodes'] = data['1']
df['status'] = data['1.1']
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df['status'].value_counts()
```

```python
df.isnull().sum()
```

```python
sns.heatmap(df.isnull(),yticklabels=False)
```

```python
sns.set_style('ticks')
sns.countplot(x='status',data=df)
```

```python
sns.set_style('darkgrid')
sns.countplot(x='nodes',data=df)
```

```python
sns.set_style('whitegrid')
sns.countplot(x='year',data=df)
```

**Bivariate Analysis**

```python
sns.FacetGrid(df,hue = 'status').map(plt.scatter,'age','year').add_legend()
```

```python
sns.FacetGrid(df,hue = 'status').map(plt.scatter,'age','nodes').add_legend()
```

Pair plot(used low dimensional data)

```python
sns. set(style="ticks")
sns.pairplot(df,hue='status',vars=['age','year','nodes'],palette='tab10')
```

Univariate

```python
live = df.loc[df['status']==1]
dead = df.loc[df['status']==2]
plt.plot(live['age'],np.zeros_like(live['age']),'o')
plt.plot(dead['age'],np.zeros_like(dead['age']),'o')
plt.show()
```

```python
sns.FacetGrid(df,hue='status').map(sns.displot,'year').add_legend()
```
