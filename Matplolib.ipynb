Plotting using matplotlib library

```python
import matplotlib.pyplot as plt
plt.plot([1,2,3],[4,45,6])
plt.title('Chart')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
x = [10,25,30]
y = [3,5,6]
x1 = [13,15,16]
y1 = [22,44,65]
plt.plot(x,y,'r',label = 'line-one',linewidth = 3)
plt.plot(x1,y1,'b',label = 'line-two',linewidth = 1)
plt.show()
```

```python
x = [10,25,30]
y = [3,5,6]
x1 = [13,15,16]
y1 = [22,44,65]
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Garph')
plt.plot(x,y,'r',label = 'line-one',linewidth = 3)
plt.plot(x1,y1,'b',label = 'line-two',linewidth = 1)
plt.grid()
plt.legend()
plt.show()
```

**Histogram**

# 3d plot

```python
import pandas as pd
```

```python
data = pd.read_csv('/content/data.csv')
```

```python
data.head()
```

```python
data['diagnosis'].value_counts()
```

```python
for diagnosis,subset in data.groupby('diagnosis'):
  plt.scatter(subset['radius_mean'],subset['perimeter_mean'],label='diagnosis')
plt.legend()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
for diagnosis,subset in data.groupby('diagnosis'):
  plt.scatter(subset['radius_mean'],subset['perimeter_mean'],subset['area_mean'],label='diagnosis')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.legend()
```

```python
import matplotlib.pyplot as plt

days = [1,2,3,4,5]

sleeping =[7,8,6,11,7]

eating = [2,3,4,3,2]

working =[7,8,7,2,2]

playing = [8,5,7,8,13]

plt.plot([],[],color='m', label='Sleeping', linewidth=5)

plt.plot([],[],color='c', label='Eating', linewidth=5)

plt.stackplot(days, sleeping,eating,working,playing,colors=['m','c','r','k'])

plt.xlabel('x')

plt.ylabel('y')

plt.title('Stack Plot')

plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
slices = [2,4,6,12]
activities = ['sleeping','eating','working','playing']
activitie = ['sleeping','eating','working','playing']
cols = ['c','m','r','b']
plt.pie(slices,labels=activities,colors=cols,startangle=180 )
plt.show()
plt.pie(slices,labels=activitie,colors=cols,startangle=30  )
plt.title('Pie Plot')
plt.show() # we can observe change in start angle
```

```python
import numpy as np

import matplotlib.pyplot as plt
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)
t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)
plt.subplot(211)  #no of plots, row wise or column wise, 1 2 plot
plt.plot(t1, f(t1), 'bo', t2, f(t2))
plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2))
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)
t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)
plt.subplot(321)  #no of plots, row wise or column wise, 1 2 plot
plt.plot(t1, f(t1),'bo')
plt.subplot(322)
plt.plot(t2, np.cos(2*np.pi*t2))
plt.subplot(323)
plt.plot(t2, np.cos(2*np.pi*t2))

plt.show()
```
